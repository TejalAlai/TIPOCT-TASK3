```python
#INFINITE WHILE LOOP FOR USER INPUT

while True:
    
    #ASK USER IF THEY WANTS TO ACCESS LIBRARY OR NOT
    
    print("\n\nIf u wan to get in a library >> enter yes ")
    print("\nIf u wan to terminate from a library >> enter exit ")
    choice = input("\n\nEnter your choice"+" yes or exit >> ")
    x = str.lower(choice)
    print("\nUser entered >> "+str(x))
    if x == "yes":
        
#CLASS NAMED LIBRARY
        class Library:
            
            def init(self):
                pass
#DEFINING A FUNCTION FOR DISPLAYING BOOKS AVAILABLE IN LIBRARY   

            def book_display(self):
                global booklist
                booklist = ["TOC","DBMS","E&M","INTRODUCTION TO RESEARCH","CP"]
                print("\nBooks available in library >> "+str(booklist))
                
#DEFINING A FUNCTION FOR LENDING BOOKS AVAILABLE IN LIBRARY         

            def lend_book(self):
                global book_name
                global lender
                book_name = input("\nEnter name of book you want to get >> ")
                lender = input("\nEnter your name >> ")
                
                if book_name in booklist:
                    booklist.remove(book_name)

                else:
                    pass
                print("\nBooks available in library after lender receives book >> "+str(booklist))
                
                global my_book
                my_book = book_name
                
#DEFINING A FUNCTION FOR ADDING NEW BOOK IN LIBRARY  

            def add_book(self):
                new_book_name = input("\nName of newly added book >> ")
                booklist.append(new_book_name)

                print("\nAfter addition of new book, books  avilable in library are >> "+ str(booklist))
#DEFINING A FUNCTION FOR RETURN BOOK 

            def return_book(self,return_book_name):
                self.return_book_name = return_book_name
                booklist.append(return_book_name)
                print("\nAfter lender submit a book, books available in library are >> "+ str(booklist))
                print("\nYou have successfully lend a book\n--------------------------***--------------------------\n")
                


        obj = Library()
        obj.book_display()
        obj.lend_book()
        obj.add_book()
        obj.return_book(str(my_book))
        
    else:
        exit()
        
```
